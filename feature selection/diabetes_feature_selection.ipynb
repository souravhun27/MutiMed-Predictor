```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
median_of_skin_thickness = diabetes_dataset['SkinThickness'].median()
median_of_insulin = diabetes_dataset['Insulin'].median()
print("median_of_skin_thickness : ",median_of_skin_thickness)
print("median_of_insulin : ",median_of_insulin)
```

```python
# Separating the data and labels, and dropping 'Skin Thickness' and 'Insulin', 'BloodPressure' columns
X = diabetes_dataset.drop(columns=['Outcome'], axis=1)
Y = diabetes_dataset['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

feature selection by RFE

```python
# Apply RFE to rank features
rfe = RFE(estimator=classifier, n_features_to_select=5)  # Select top 5 features
rfe.fit(X_train, Y_train)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

# Get the ranking of features
feature_ranking = pd.DataFrame({
    'Feature': X.columns,
    'Ranking': rfe.ranking_
}).sort_values(by='Ranking')

print("Feature Rankings:\n", feature_ranking)
```

```python
# Select the top-ranked features
top_features = feature_ranking[feature_ranking['Ranking'] == 1]['Feature'].values
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]


# Train SVM on the selected top features
svm_model_top = SVC(kernel='linear') # Create a new SVC instance
svm_model_top.fit(X_train_top, Y_train)


# Evaluate the model on training and test data
train_accuracy = accuracy_score(Y_train, svm_model_top.predict(X_train_top))
test_accuracy = accuracy_score(Y_test, svm_model_top.predict(X_test_top))
conf_matrix = confusion_matrix(Y_test, svm_model_top.predict(X_test_top))

print("Training Accuracy with Top Features:", train_accuracy)
print("Test Accuracy with Top Features:", test_accuracy)
print("Confusion Matrix on Test Data:\n", conf_matrix)
```

Model Evaluation after feature selection

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```
