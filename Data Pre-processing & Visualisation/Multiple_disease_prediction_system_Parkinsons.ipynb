Data visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Set up Seaborn style
sns.set(style="whitegrid")
# Distribution Plots
# Distribution of each numeric feature
for column in parkinsons_data.columns[1:]:  # Skip 'name' column
    plt.figure(figsize=(8, 4))
    sns.histplot(parkinsons_data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Box plot for each feature by 'status'
for column in parkinsons_data.columns[1:-1]:  # Skip 'name' and 'status'
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='status', y=column, data=parkinsons_data)
    plt.title(f'Box plot of {column} by status')
    plt.show()
```

```python
# Pair plot of selected features with 'status' as hue
selected_features = parkinsons_data.columns[1:6]  # Select the first few numeric columns for simplicity
sns.pairplot(parkinsons_data[selected_features.insert(0, 'status')], hue='status', diag_kind='kde', palette='Set1')
plt.show()
```

```python
# Scatter plot matrix for selected features
pd.plotting.scatter_matrix(parkinsons_data[selected_features], figsize=(14, 10), diagonal='kde', color='blue')
plt.suptitle("Scatter Plot Matrix for Selected Features")
plt.show()
```

Separating the features & Target

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)


if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("The Person has Parkinsons")
```
