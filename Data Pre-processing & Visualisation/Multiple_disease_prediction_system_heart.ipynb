Importing the Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('/content/heart.csv')
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

Data Visualization

```python
# 1. Distribution of Numerical Features
plt.figure(figsize=(12, 10))
for i, column in enumerate(['age', 'trestbps', 'chol', 'thalach', 'oldpeak']):
    plt.subplot(3, 2, i + 1)
    sns.histplot(heart_data[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# 2. Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = heart_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# 3. Pairplot to Explore Feature Interactions (colored by 'target')
sns.pairplot(heart_data, hue='target', palette='coolwarm', diag_kind='kde')
plt.suptitle("Pairplot of Features Colored by Target", y=1.02)
plt.show()
```

```python
# 4. Boxplots of Key Features by Target
plt.figure(figsize=(12, 8))
for i, column in enumerate(['age', 'trestbps', 'chol', 'thalach', 'oldpeak']):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(x='target', y=column, data=heart_data, palette='coolwarm')
    plt.title(f'{column} by Target')
plt.tight_layout()
plt.show()
```

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
conf_matrix = confusion_matrix(Y_test, X_test_prediction)

# Visualize the confusion matrix using seaborn
plt.figure(figsize=(4, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])

plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.title("Confusion Matrix")
plt.show()
```

Building a Predictive System

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
